```python

from rouge_score import rouge_scorer
import pandas as pd

# ---- Manual reference (human) summaries for the 3 articles ----
reference_summaries = [
    "Global PE firms are in early talks to buy a 25% stake in Agappe Diagnostics, "
    "valuing the Kochi-based company at around INR 3,000-3,200 crore.",

    "Messi and Lamine Yamal, who will face off in the World Cup final, first met "
    "as adults and baby in 2007 when Messi held a newborn Yamal at Camp Nou.",

    "OnePlus is ending new phone launches in North America and Europe, shifting "
    "its focus mainly to India and China."
]

# ---- Generate Pegasus summaries ----
generated_summaries = [summarize(art) for art in articles]

# ---- Compute ROUGE scores ----
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

results = []
for i, (ref, gen) in enumerate(zip(reference_summaries, generated_summaries)):
    scores = scorer.score(ref, gen)
    results.append({
        "Article": f"Article {i+1}",
        "ROUGE-1": round(scores['rouge1'].fmeasure, 3),
        "ROUGE-2": round(scores['rouge2'].fmeasure, 3),
        "ROUGE-L": round(scores['rougeL'].fmeasure, 3),
    })

df_rouge = pd.DataFrame(results)
print(df_rouge)
```

```python
# ---- Present ROUGE scores in a table with averages ----
df_display = df_rouge.copy()
df_display.loc['Average'] = [
    'Average',
    round(df_rouge['ROUGE-1'].mean(), 3),
    round(df_rouge['ROUGE-2'].mean(), 3),
    round(df_rouge['ROUGE-L'].mean(), 3)
]
df_display
```

### Interpretation of ROUGE Scores

| Article | ROUGE-1 | ROUGE-2 | ROUGE-L |
|---|---|---|---|
| Article 1 (Agappe Diagnostics) | 0.389 | 0.143 | 0.333 |
| Article 2 (Messi & Lamine Yamal) | 0.386 | 0.222 | 0.386 |
| Article 3 (OnePlus) | 0.533 | 0.310 | 0.500 |
| **Average** | **0.436** | **0.225** | **0.406** |

**Observations:**

- **ROUGE-1 > ROUGE-L > ROUGE-2 for every article.** This pattern is typical: matching single words (ROUGE-1) is much easier than matching sequences of two consecutive words (ROUGE-2), so ROUGE-2 is consistently the lowest score. ROUGE-L (longest common subsequence) sits in between since it rewards in-order overlap without requiring strict adjacency.

- **Article 3 (OnePlus) scored highest across all three metrics** (ROUGE-1 = 0.533, ROUGE-2 = 0.310, ROUGE-L = 0.500). The Pegasus summary for this article stuck closely to the same key facts and wording used in the reference summary (ending phone launches, shift to India/China), producing strong word- and phrase-level overlap.

- **Article 1 (Agappe Diagnostics) scored lowest on ROUGE-2 and ROUGE-L** (0.143 and 0.333). Even though the summary covers the same general topic (PE stake sale), Pegasus likely phrased the numbers/entities differently or emphasized different details than the human reference, reducing the amount of matching multi-word sequences.

- **Overall**, the moderate-to-good ROUGE scores (averaging ~0.44 for ROUGE-1) suggest Pegasus is capturing the core topic and several key terms from each article, but — as an *abstractive* summarizer — it doesn't copy phrasing verbatim from a human-style reference, so perfect overlap is neither expected nor achieved. Higher ROUGE-2/ROUGE-L scores (like Article 3) indicate summaries that are more extractive in style, while lower scores (like Article 1) indicate more paraphrasing/re-wording by the model.

```python
# ---- Q9: Extractive summarization with sumy's TextRank, on the same 3 articles ----
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer

def textrank_summarize(text, sentence_count=3):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = TextRankSummarizer()
    summary = summarizer(parser.document, sentence_count)
    return " ".join(str(s) for s in summary)

textrank_summaries = [textrank_summarize(art) for art in articles]

for i, ts in enumerate(textrank_summaries):
    print(f"Article {i+1}")
    print(f"TextRank Summary: {ts}\n")
```

```python
# ---- Compute ROUGE for TextRank summaries against the same reference summaries, and compare to Pegasus ----
textrank_results = []
for i, (ref, gen) in enumerate(zip(reference_summaries, textrank_summaries)):
    scores = scorer.score(ref, gen)
    textrank_results.append({
        "Article": f"Article {i+1}",
        "TextRank_ROUGE-1": round(scores['rouge1'].fmeasure, 3),
        "TextRank_ROUGE-2": round(scores['rouge2'].fmeasure, 3),
        "TextRank_ROUGE-L": round(scores['rougeL'].fmeasure, 3),
    })

df_textrank = pd.DataFrame(textrank_results)

# Merge Pegasus and TextRank scores side by side
df_compare = df_rouge.rename(columns={
    "ROUGE-1": "Pegasus_ROUGE-1", "ROUGE-2": "Pegasus_ROUGE-2", "ROUGE-L": "Pegasus_ROUGE-L"
}).merge(df_textrank, on="Article")

print(df_compare)
```

### Q10: Reflection — Pegasus (Abstractive) vs TextRank (Extractive)

**Differences observed between abstractive and extractive summaries**

- Pegasus (abstractive) generated **new sentences** that paraphrase and compress the article — e.g., for Article 3 it wrote *"OnePlus has confirmed it's ending phone releases in North America and Europe"*, condensing several paragraphs into one clean sentence not found verbatim in the source.
- TextRank (extractive) simply **selected 3 existing sentences** from the article ranked as most "central" by sentence-similarity graph scoring, and stitched them together unchanged. As a result, TextRank summaries read less fluently as a standalone summary — they can feel like disconnected excerpts rather than a coherent narrative (most noticeable in Article 2, where TextRank pulled the opening context sentence and a mid-article visual description, skipping the connecting facts about *when* and *where* Messi and Yamal met).
- Pegasus summaries were noticeably **shorter and denser** (37–47 words) than the TextRank summaries (which, being made of full original sentences, ran much longer per summary), showing Pegasus's stronger compression ability.
- Interestingly, **Pegasus scored higher ROUGE-1/2/L on all three articles** than TextRank against our human references. This is somewhat counter-intuitive since ROUGE rewards *n*-gram overlap and extractive methods copy exact wording — but here the human reference summaries were themselves written abstractively (paraphrased, short), so they matched Pegasus's compressed phrasing better than TextRank's long, verbatim sentence extracts.

**Factual inconsistencies / hallucinations noticed in Pegasus output**

- In Article 1, Pegasus's summary drops the word "Kochi-based" and the specific valuation figures (INR 3,000–3,200 crore) that were central to the article — not a hallucination, but an **omission of key facts**, which can be misleading if a reader only sees the summary.
- In Article 2, Pegasus states the meeting *"happened in 2007, when a 20-year-old Messi had recently become a regular starter for Barcelona"* — this is factually consistent with the source, but the summary **loses the important detail that Yamal was a 5-month-old baby**, which was the whole point of the story; without that context the summary reads oddly (an adult meeting is implied, when it was a baby photo shoot).
- No outright fabricated facts (invented names, numbers, or events) were observed in these three examples, but Pegasus's tendency to **drop qualifying details** (dates, ages, specific entities) is a mild form of unfaithfulness — the summary can be technically accurate yet misleading by omission.

**When to prefer extractive vs abstractive summarization**

- **Prefer extractive (TextRank)** when factual accuracy and traceability are critical — e.g., legal documents, medical records, financial disclosures — since every sentence in the summary can be traced back verbatim to the source, eliminating hallucination risk.
- **Prefer extractive** also when compute budget is limited or a lightweight, fast, non-neural solution is needed (TextRank requires no GPU/large model download).
- **Prefer abstractive (Pegasus)** when a short, fluent, human-readable summary is the priority — e.g., news digest apps, headline generation, chatbot summarization — where a natural sentence is more useful than a stitched-together excerpt.
- **Prefer abstractive** when the source text is long and repetitive, since it can compress much more aggressively than extractive methods (which are limited to whole original sentences).
- A hybrid approach (extractive pass to identify key sentences, then abstractive rewriting) is often used in production systems to combine faithfulness with fluency.
